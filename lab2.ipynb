```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D

# Generating sample n-dimensional data
np.random.seed(42)
num_samples = 100
num_dimensions = 3

data = np.random.randn(num_samples, num_dimensions)
df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3'])

# Scatter plot for 2D data
sns.scatterplot(x='Feature1', y='Feature2', data=df)
plt.title('Scatter Plot for 2D Data')
plt.show()
```

```python
# Box plot for each feature
sns.boxplot(data=df)
plt.title('Box Plot for Each Feature')
plt.show()
```

```python
# Heat map for correlation matrix
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heat Map for Correlation Matrix')
plt.show()
```

```python
# Contour plot for 2D data
sns.kdeplot(x='Feature1', y='Feature2', data=df, fill=True, cmap='viridis', levels=10)
plt.title('Contour Plot for 2D Data')
plt.show()
```

```python
# 3D surface plot for 3D data
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['Feature1'], df['Feature2'], df['Feature3'], c='b', marker='o')
ax.set_xlabel('Feature1')
ax.set_ylabel('Feature2')
ax.set_zlabel('Feature3')
ax.set_title('3D Scatter Plot')
plt.show()
```
