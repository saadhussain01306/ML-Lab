```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ToyotaCorolla.csv")

x = data["KM"]
y = data["Price"]
z = data["Weight"]

plt.scatter(x, y, c=z, cmap='summer', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label("KM/Price")

plt.title("Price of Toyota Corolla")
plt.xlabel("KM")
plt.ylabel("Price")
plt.tight_layout()

plt.show() 
```

```python
# box plot
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ToyotaCorolla.csv")

# Use the correct column names with proper case sensitivity
plt.boxplot([data["Price"], data["HP"], data["KM"]])
plt.xticks([1, 2, 3], ["Price", "HP", "KM"])
plt.show()
```

```python
#Heat Maps

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv("ToyotaCorolla.csv")

# Selecting specific columns for the correlation matrix
selected_columns = ["Price", "KM", "Doors", "Weight"]
correlation_matrix = data[selected_columns].corr()

# Creating a heatmap using seaborn
sns.heatmap(correlation_matrix, cmap="jet", annot=True)

plt.title("Correlation Heatmap")
plt.show()
```

```python
# contour plot
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv("ToyotaCorolla.csv")

x = data["KM"]
y = data["Weight"]
z = data["Price"]

plt.tricontour(x, y, z, levels=10, cmap="jet")
plt.colorbar(label="Price")
plt.xlabel("Weight")
plt.title("Contour Plot")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("ToyotaCorolla.csv")

x = data["KM"]
y = data["Doors"]
z = data["Price"]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot_trisurf(x, y, z, cmap="jet")

ax.set_title("3D Surface Plot")
ax.set_xlabel("KM")
ax.set_ylabel("Doors")
ax.set_zlabel("Price")

plt.show()
```
